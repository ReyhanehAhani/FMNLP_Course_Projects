```python
import csv

# Read funny descriptions from the funny file
funny_descriptions = []
with open('/content/FlickrStyle_v0.9/humor/funny_train.txt', 'r', encoding='latin-1') as funny_file:
    funny_descriptions = funny_file.readlines()

# Read image names from the train.p file
image_names = []
with open('/content/FlickrStyle_v0.9/humor/train.p', 'r', encoding='latin-1') as train_file:
    lines = train_file.readlines()
    for line in lines:
        if line.startswith('aV') or line.startswith('V'):
            image_name = line.split('_')[0][1:]
            image_names.append(image_name)

# Create a dictionary to store image names and corresponding descriptions
image_dict = {}
for i, image_name in enumerate(image_names):
    image_dict[image_name] = {
        'funny_description': funny_descriptions[i].strip(),
        'romantic_description': None  # Initialize as None, to be filled later
    }

# Read romantic descriptions from the romantic file
with open('/content/FlickrStyle_v0.9/romantic/romantic_train.txt', 'r', encoding='latin-1') as romantic_file:
    romantic_descriptions = romantic_file.readlines()

# Fill in the romantic descriptions in the dictionary
for i, image_name in enumerate(image_names):
    image_dict[image_name]['romantic_description'] = romantic_descriptions[i].strip()

# Write the data to a CSV file
with open('output.csv', 'w', newline='') as csvfile:
    fieldnames = ['Image Name', 'Funny Description', 'Romantic Description']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for image_name, descriptions in image_dict.items():
        writer.writerow({
            'Image Name': f'{image_name}.jpg',
            'Funny Description': descriptions['funny_description'],
            'Romantic Description': descriptions['romantic_description']
        })

print("CSV file created successfully.")
```

```python
import os
import shutil
import pandas as pd

def copy_images(source_folder, destination_folder, csv_file):
    # Load CSV file into a DataFrame
    df = pd.read_csv(csv_file)

    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    for index, row in df.iterrows():
        image_name = row['Image Name']
        source_path = os.path.join(source_folder, image_name)
        destination_path = os.path.join(destination_folder, image_name)

        if os.path.exists(source_path):
            shutil.copy(source_path, destination_path)
            print(f"Image '{image_name}' copied successfully.")
        else:
            print(f"Image '{image_name}' not found in the source folder.")

source_folder_path = '/content/flickr30k_images/flickr30k_images'
destination_folder_path = '/content/FMNLP_dataset/flickerstyle/images'
csv_file_path = '/content/output_one_neutral.csv'

copy_images(source_folder_path, destination_folder_path, csv_file_path)
```

```python
from transformers import AutoProcessor, BlipForConditionalGeneration

processor = AutoProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
```

```python
import torch
```

```python
for name, parameter in model.named_parameters():
    if 'text_decoder' in name:
        parameter.requires_grad = True
    else:
        parameter.requires_grad = True #False
```

```python
funny_token = "[FUNNY]"
romantic_token = "[ROMANTIC]"
neutral_token = "[NEUTRAL]"
```

```python
tokenizer = processor.tokenizer
bert = model.text_decoder.bert

special_tokens_dict = {'additional_special_tokens': [funny_token, romantic_token, neutral_token]}
num_added_toks = tokenizer.add_special_tokens(special_tokens_dict)
bert.resize_token_embeddings(len(tokenizer))
```

```python
model.text_decoder.cls.predictions.decoder = torch.nn.Linear(in_features=768, out_features=len(tokenizer), bias=True)
```

```python
for i in (pbar := tqdm(range(0, len(test_df)))):

  text_neutral = df.iloc[i, 4] #
  text_funny = df.iloc[i, 1]
  text_romantic = df.iloc[i, 2]

  text_funny = neutral_token + ':' + text_neutral + funny_token + ':' + text_funny
  text_neutral_ref = neutral_token + ':' + text_neutral + neutral_token + ':' + text_neutral
  text_romantic = neutral_token + ':' + text_neutral + romantic_token + ':' + text_romantic


  images = Image.open(os.path.join('/content/FMNLP_dataset/flickerstyle/images', df.iloc[i, 0]))

  '''
  funny_pre = processor(text=funny_token + text_neutral, images=images, return_tensors="pt", truncation=True).to(DEVICE)
  romantic_pre = processor(text=romantic_token + text_neutral, images=images, return_tensors="pt", truncation=True).to(DEVICE)
  neutral_pre = processor(text=neutral_token + text_neutral, images=images, return_tensors="pt", truncation=True).to(DEVICE)

  funny_out = processor.decode(model.generate(**funny_pre, max_new_tokens=100)[0], skip_special_tokens=True)
  romantic_out = processor.decode(model.generate(**romantic_pre, max_new_tokens=100)[0], skip_special_tokens=True)
  neutral_out = processor.decode(model.generate(**neutral_pre, max_new_tokens=100)[0], skip_special_tokens=True)
  '''

  pre = processor(text=(neutral_token + text_neutral + funny_token, neutral_token + text_neutral + romantic_token, neutral_token + ':' + text_neutral + neutral_token + ':'), images=(images, images, images), return_tensors="pt", truncation=True).to(DEVICE)
  results = model.generate(**pre)
  decode = lambda x: processor.decode(x)
  funny_out = decode(results[0])
  romantic_out = decode(results[1])
  neutral_out = decode(results[2])

  funny_bleu += sentence_bleu([text_funny.split()], funny_out.split())
  romantic_bleu += sentence_bleu([text_romantic.split()], romantic_out.split())
  neutral_bleu += sentence_bleu([text_neutral_ref.split()], neutral_out.split())

  funny_rouge = rouge.get_scores(funny_out, text_funny)
  funny_rouge_1 += funny_rouge[0]['rouge-1']['f']
  funny_rouge_2 += funny_rouge[0]['rouge-2']['f']
  funny_rouge_l += funny_rouge[0]['rouge-l']['f']

  romantic_rouge = rouge.get_scores(romantic_out, text_romantic)
  romantic_rouge_1 += romantic_rouge[0]['rouge-1']['f']
  romantic_rouge_2 += romantic_rouge[0]['rouge-2']['f']
  romantic_rouge_l += romantic_rouge[0]['rouge-l']['f']

  neutral_rouge = rouge.get_scores(neutral_out, text_neutral_ref)
  neutral_rouge_1 += neutral_rouge[0]['rouge-1']['f']
  neutral_rouge_2 += neutral_rouge[0]['rouge-2']['f']
  neutral_rouge_l += neutral_rouge[0]['rouge-l']['f']

  funny_meteor += meteor_score.meteor_score([text_funny.split()], funny_out.split())
  romantic_meteor += meteor_score.meteor_score([text_romantic.split()], romantic_out.split())
  neutral_meteor += meteor_score.meteor_score([text_neutral_ref.split()], neutral_out.split())
```
